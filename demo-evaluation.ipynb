# Evaluation

```python
model = 'batterydata/batterybert-cased-squad-v1'
from transformers import pipeline
qa = pipeline('question-answering',model, use_auth_token=True)
```

```python
import pandas as pd
property_name = 'voltage'
loc = 'tests/evaluation/{}.csv'.format(property_name)
tests = pd.read_csv(loc)
```

```python
tests
```

```python
names = tests['Name'].to_list()
sentences = tests['Text'].to_list()
values = tests['Value'].to_list()
```

```python
import unicodedata
for i in range(20):
    name = names[i]
    text = sentences[i]
    value = values[i]
    print(name, value)
    print(qa("What's the {} of {}?".format(property_name, name), unicodedata.normalize("NFKD", text)))
    print(text)
    print('\n')
```

```python
import pandas as pd
property_name = 'capacity'
loc = 'tests/evaluation/{}.csv'.format(property_name)
tests = pd.read_csv(loc)
names = tests['Name'].to_list()
sentences = tests['Text'].to_list()
values = tests['Value'].to_list()

from batterydataextractor.doc import Document
import unicodedata
for i in range(20):
    name = names[i]
    text = sentences[i]
    value = values[i]
    print(name, value)
    doc = Document(text)
    print(text)
    # TODO: Need to get plural manually now.
    doc.add_models_by_names(["capacity", "capacities"])
    record = doc.records
    for r in record:
        print(r.serialize())
    print('\n')
```
